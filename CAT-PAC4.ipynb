Programació per a la ciència de dades - PAC4
============================

En aquest Notebook trobareu l'exercici que suposa la quarta i darrera activitat d'avaluació continuada (PAC) de l'assignatura. Aquesta PAC intenta presentar-vos un petit projecte en el qual heu de resoldre diferents exercicis, que engloba molts dels conceptes coberts durant l'assignatura.

L'objectiu d'aquest exercici serà desenvolupar un projecte Python fora de l'entorn de Notebooks, que ens permeti resoldre el problema donat.
Treballareu amb arxius Python plans `.py`. El projecte haurà d'incloure el corresponent codi organitzat lògicament (separat per mòduls, organitzats per funcionalitat,...), la documentació del codi (*docstrings*) i tests. A més, haureu d'incloure els corresponents arxius de documentació d'alt nivell (`README`) així com els arxius de llicència i dependències (`requirements.txt`) comentats a la teoria.

Fer un `setup.py` és opcional, però si es fa es valorarà positivament de cara a la nota de la pràctica i del curs.

## README, LICENCE, requirements.txt

* **requirements.txt**: fitxer de requeriments que contingui la llista de llibreries necessàries per executar el codi. Han de ser les llibreries pròpies d'aquest projecte (no cal incloure, tot i que es poden incloure, les llibreries accessòries per executar els tests, linter, documentació).

## Modularitat

El codi ha de ser modular, tant en les funcions que es defineixen (programació estructurada), com en l'organització del codi en mòduls. Una possibilitat és separar cada exercici en un fitxer. Un altre alumne preferirà que els fitxers de mòduls siguin funcionalitats. Típicament hi haurà el fitxer *main.py*, i aquest fitxer importarà els mòduls i executarà els diferents exercicis. La idea és que tota la funcionalitat estigui en els mòduls, i el fitxer *main.py* serveix com a punt d'entrada per llegir les opcions i executar els mòduls/exercicis.

## Opcions a la línia de comandes

Pots utilitzar directament la llibreria *sys*, o bé alguna altra possibilitat (*getopt* o *argparse* que és la que s'utiliza a la solució), tant se val. El que es demana és que hi hagi l'opció d'ajuda (-h, --help), o bé executar els exercicis individualment (-ex num). Si passem *-ex 5* pressuposa que s'executa de l'exercici 1 al 5, doncs els exercicis són progressius. Si no passem arguments, s'executen tots els exercicis. Documentar al *README*.
